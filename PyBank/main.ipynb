```python
import os
import csv

rows = []
mounth = []
dates = []
count_months = 0
current_month_profit_loss = 0
net_profit_loss = 0
x = 0 
avg = 0
y = 0 

budget_data = ("C:/Users/alisa/GIT/python-challenge/PyBank/budget_data.csv")


with open(budget_data) as csvfile:

    csv_reader = csv.reader(csvfile, delimiter=",")

    csv_header = next(csvfile)
    
    for row in csv_reader:
        first_column = row[0].split('-')
        dates.append(first_column)
        second_column = int(row[1])
        profit_losses.append(second_column)

        count_months += 1
        
        rows.append(row) 
        mounth.append(row)
                         
#To calculate net income, take the gross income — the total amount of money earned — then subtract expenses, such as taxes and interest payments.
       # current_month_profit_loss = second_column
        last_month_profit_loss_date = str(row[0])
        net_profit_loss += second_column
#----------------------Average Change        
    x = int(rows[(count_months - 1)][1])
    y = int(rows[0][1])
    z = round(((x) - (y)) / (count_months - 1))
    
#------------------Greatest Increase & Decrease  in Profits

    difchanges = []
    
    for i in range(len(profit_losses)-1):
        difference = profit_losses[i+1]-profit_losses[i]
        difchanges.append(difference) 
        
   
    highest_change = max(difchanges)
    lowest_change = min(difchanges)
    
    h_d_index =  difchanges.index(highest_change)
    h_d_month = mounth[h_d_index+1][0]
    
    L_d_index =  difchanges.index(lowest_change)
    L_d_month = mounth[L_d_index+1][0]   

#----------------summery report---------------    
print("Financial Analysis")
print("----------------------------")  
print(f"Total Months:{count_months}")
print("Total:"+str(net_profit_loss))
print("Average Change:"+str(z))
#print("Greatest Increase in Profits:"+str(highest_change))
#print("Greatest Decrease in Profits:"+str(lowest_change))
print("Greatest Increase in Profits:"+str(h_d_month)+"  $"+str(highest_change))
print("Greatest Decrease in Profits:"+str(L_d_month)+"  $"+str(lowest_change))

#----------------write summery to analaysis file---------------
analysis = open ("C:/Users/alisa/GIT/python-challenge/analysis-budget_file.txt", 'w')
analysis.write("Financial Analysis"+ "\n")
analysis.write("----------------------------"+ "\n")  
analysis.write(f"Total Months:{count_months}"+ "\n")
analysis.write("Total:"+str(net_profit_loss)+ "\n")
analysis.write("Average Change:"+str(z)+ "\n")
#print("Greatest Increase in Profits:"+str(highest_change))
#print("Greatest Decrease in Profits:"+str(lowest_change))
analysis.write("Greatest Increase in Profits:"+str(h_d_month)+"  $"+str(highest_change)+ "\n")
analysis.write("Greatest Decrease in Profits:"+str(L_d_month)+"  $"+str(lowest_change)+ "\n")

analysis.close()
    
                 
```
